# データ分析

今回のデータ（日付、時間、商品カテゴリ、商品ID、単価、数量）に基づいた分析フレームワークを以下の項目に分けて検討します。それぞれが適切かどうかを判断し、分析手法や意義を説明しながら進めます。

1. **全期間・イベントに基づいたデータ分析**
**概要**
- **全期間の売上や数量の推移を確認**し、コロナなどの特定のイベント期間における傾向を捉える。
- 各イベント期間（例：緊急事態宣言やGoToキャンペーン）ごとに売上や数量がどのように変化したかを時系列データで分析します。

**適切かどうか**
適切です。特に、コロナ期間やその他の重要イベントの影響を定量的に評価することは、マーケティングや施策の立案に重要な役割を果たします。イベントによる変化が見られれば、今後の需要予測やイベント対策にも役立てられます。

**推奨される手法**
- **時系列分析**を用いて、売上の増減や数量の変動を確認。
- **イベント期間ごとの平均売上や数量**を集計し、イベントの影響を可視化。
- **イベント前後の比較分析**で影響を定量的に測定。

---

2. **データ分析（1変量・2変量・多変量分析）**
**概要**
- **1変量分析**: 各変数の基本的な傾向（例：数量の平均、分散、ヒストグラム）。
- **2変量分析**: 2つの変数間の関係性を確認（例：商品単価と数量の相関、時間帯と売上の関係）。
- **多変量分析**: 複数の変数を組み合わせて売上や数量に影響を与える要因を特定（例：商品カテゴリ、時間帯、単価を説明変数にした数量の回帰分析）。

**適切かどうか**
適切です。1変量分析ではデータの全体像を把握でき、2変量・多変量分析では変数間の関係を深掘りして、売上や数量に影響を与える要因を特定できます。これにより、特定の商品の価格設定や時間帯ごとの施策など、具体的な戦略を立案するための洞察を得られます。

**推奨される手法**
- **1変量分析**: 各変数の分布（ヒストグラム、基本統計量）。
- **2変量分析**: 散布図や相関係数を使い、変数間の関係性を可視化。
- **多変量分析**: 回帰分析やクラスタリングを使って、複数の要因が売上にどう影響しているかを分析。

---

3. **モニタリング（KGI・KPIのダッシュボード）**
**概要**
- **KGI（Key Goal Indicator）**として、全体売上の増加や目標達成率を設定。
- **KPI（Key Performance Indicator）**として、商品カテゴリごとの売上、月・日・時別の売上割合、イベント期間ごとの売上推移をモニタリングする。

**適切かどうか**
非常に適切です。ビジネスにおける目標を設定し、その達成度をリアルタイムでモニタリングすることは、戦略の調整に役立ちます。特にKGIとして売上目標を設定し、その達成に向けたKPI（カテゴリごとの売上や時間帯ごとの販売割合）を追跡することで、戦略の効果を評価できます。

**推奨される手法**  
- **ダッシュボードの作成**: PowerBIやTableau、またはPythonのPlotlyやDashを使用して、リアルタイムでKPIを視覚化する。
- **KPIモニタリング**: 月別・日別・時間別の売上、商品カテゴリごとの割合、コロナ期間の売上推移をダッシュボードに表示。
- **アラート機能**: KPIが設定した目標値を下回った場合にアラートを出す仕組みを組み込む。

---

4. **売上の時系列分析**
**概要**  
- 売上の時系列データを使い、全体の売上推移や季節的な変動、イベントによる影響を確認します。
- 特に、月・日・時間ごとの売上割合を分析し、ピーク時間帯やオフピーク時間帯の特定。

**適切かどうか**  
適切です。売上の時系列データを分析することで、季節性やトレンド、イベントの影響を正確に捉えられます。これにより、予測モデルの構築や施策の評価が可能です。

**推奨される手法**  
- **時系列予測モデル**: ARIMAやLSTMなどを用いて、売上の予測モデルを構築。
- **トレンド分析**: 季節性や傾向を視覚化し、将来の売上予測に活用。

---

5. **商品カテゴリの割合分析**
**概要**  
- 商品カテゴリごとの売上や数量の割合を確認し、どのカテゴリが全体売上にどの程度寄与しているかを把握します。

**適切かどうか**  
適切です。商品カテゴリごとの売上や数量の割合を確認することで、人気のある商品や販売の中心となるカテゴリを把握できます。これに基づいて在庫管理やマーケティング施策の強化が可能です。

**推奨される手法**  
- **円グラフや棒グラフ**で各カテゴリの割合を可視化し、売上貢献度を視覚的に把握。
- **カテゴリごとの推移**: 時間とともにカテゴリごとの売上がどう変化したかも確認する。

---

6. **月・日・時別の割合分析**
**概要**  
- 月別・日別・時間別の売上割合を確認し、売上がどの時間帯や曜日、月に集中しているかを特定します。

**適切かどうか**  
非常に適切です。売上がどのタイミングで増加しているかを知ることで、プロモーションのタイミングや特定時間帯での施策を考える際に役立ちます。これにより、売上を最大化するための施策を打ち出せます。

**推奨される手法**  
- **ヒートマップ**や**棒グラフ**で月別・日別・時間別の売上割合を可視化し、ピーク時間帯やオフピーク時間帯を特定。
- **曜日や季節ごとの傾向**も併せて分析し、特定の時期に集中する需要に備える。

---

**総合的な判断**  

全体として、これらの分析フレームワークは適切です。特に、売上の時系列分析や商品カテゴリごとの割合分析、そしてKPIに基づいたモニタリングは、ビジネスの意思決定に役立つ強力なツールです。さらに、季節性や時間帯別の需要を把握することで、将来的なプロモーションや価格戦略にも役立つ情報が得られます。

次のステップとしては、これらのデータ分析をシステマチックに行い、リアルタイムでモニタリング可能なダッシュボードを構築することが推奨されます。また、機械学習を用いた予測モデルを導入することで、未来の売上をより精度高く予測し、施策の効果を検証することができるでしょう。

### 作業手順（修正版）

---

1. **目的変数の確認（可視化）**：
   - まず、目的変数（今回の場合は「数量」）の基本的な分布や傾向を把握するために、**棒グラフ**や**ヒストグラム**を使用して全体像を視覚化します。これにより、数量の大まかな傾向を確認し、分析の方向性を決定します。

---

2. **簡単な分析（1変量・2変量分析）**：
   - **1変量分析**：数量の単独分布を確認します。
     - ヒストグラムや棒グラフを使用して、商品の**カテゴリごとの全数量**や各カテゴリの数量シェアを確認します。
     - **商品カテゴリごとの全数量**を棒グラフで示し、**全数量に対する商品カテゴリの割合**を円グラフで表示します。

   - **2変量分析**：2つの変数間の関係性をシンプルに分析します。
     - **商品単価と数量の関係**、**売上と数量の関係**を**散布図**で可視化し、傾向を確認します。
     - 商品カテゴリ別の数量に絞り、特定カテゴリが全体に占める割合やトレンドを簡潔に掴むことを目的とします。

---

3. **時系列分析**：
   - **全商品**に対して、**時間帯別・日別の売上や数量の推移**を確認し、**時系列グラフ**を使用して全体のトレンドを視覚化します。
   - これにより、どの時間帯や日に売上や数量が急増・減少しているのか、顧客の動向を確認します。特にピークとなる時間帯や日の特定が可能です。

---

4. **多変量解析**：
   - **ヒートマップ**を使用して、**商品カテゴリと時間帯ごとの数量や売上**を視覚化します。
     - ここでは、商品カテゴリごとの**特定の時間帯における販売傾向**を把握します。
     - これにより、例えば「撮影料」や「ガイド」などの高単価サービスが**特定の時間帯**にどれだけ集中して売れているかを確認し、施策の立案に活かします。

   - **相関分析**を行い、複数の変数（単価、数量、売上、時間帯など）の関係性を確認します。これにより、商品単価や数量に対する影響因子を明らかにします。

---

5. **分析から考えられる施策の立案**：
   - 分析結果から、特に需要が高い時間帯や商品カテゴリに対して、**値上げやプロモーション施策**を検討します。
   - 高単価サービス（撮影料やガイド）が需要集中する時間帯における**価格調整**や、販売促進の方向性を決定します。

---

6. **シミュレーションデータの作成**：
   - **価格上昇シミュレーション**を行い、分析結果に基づき、特定の時間帯で価格を上昇させた場合の**売上変動を予測**します。
   - 例えば、17:00〜18:00の時間帯に20%の値上げをした場合に、売上がどのように変化するかをシミュレーションします。

---

7. **モニタリングとKPI/KGI設定**：
   - **ダッシュボード**を使用して、設定した**KGI/KPI**に対して継続的にモニタリングを行います。
   - **売上の時系列データ**や、**商品カテゴリごとの売上・数量割合**をリアルタイムで確認できるようにします。
   - 施策実施後にその効果を定量的に評価し、次の施策の改善に繋げます。

---

**まとめ**  
この作業手順は、最初に基本的な数量の分析を行い、時系列や多変量分析を組み合わせることで全体のパフォーマンスと時間帯別・カテゴリ別のパフォーマンスを把握し、適切な施策の立案とモニタリングに繋げます。

```python
%%writefile my_custom_functions.py

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def calculate_total_sales(events, df):
    """
    各イベント期間中の売上を集計する関数。
    events: イベントのリスト（開始日と終了日を含む）
    df: 売上データが含まれるデータフレーム（日付がインデックス）
    """
    total_sales = []
    for event in events:
        event_period = df.loc[event['start']:event['end']]
        total_sales.append(event_period['売上'].sum())
    return total_sales

def plot_category_sales_pie(events, df):
    """
    イベント期間中の商品カテゴリ別売上割合を円グラフで表示する関数。
    events: イベントのリスト（開始日と終了日を含む）
    df: 売上データとカテゴリ情報が含まれるデータフレーム
    """
    for event in events:
        event_period = df.loc[event['start']:event['end']]
        category_sales = event_period.groupby('商品カテゴリ')['売上'].sum()
        plt.figure(figsize=(6, 6))
        plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
        plt.title(f"{event['event']}期間のカテゴリ別売上割合")
        plt.show()

def plot_sales_comparison(actual_sales, predicted_sales, events_df):
    """
    実売上と予測売上を比較するグラフを描画する関数。
    actual_sales: 実際の売上データ（日付がインデックス）
    predicted_sales: 予測された売上データ（日付がインデックス）
    events_df: イベント情報を持つデータフレーム（開始日と終了日を含む）
    """
    plt.figure(figsize=(10, 6))
    plt.plot(actual_sales.index, actual_sales, label='実際の売上', color='blue')
    plt.plot(predicted_sales.index, predicted_sales, label='予測の売上', color='green')
    for _, event in events_df.iterrows():
        plt.axvspan(event['start'], event['end'], color='gray', alpha=0.3, label=event['event'])
    plt.title('実際の売上と予測の売上の比較')
    plt.xlabel('日付')
    plt.ylabel('売上 (¥)')
    plt.legend(loc='upper left')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def label_encoding_for_columns(df, columns):
    """
    指定された列をラベルエンコーディングする関数。
    df: データフレーム
    columns: ラベルエンコーディングを行う列のリスト
    """
    label_encoders = {}
    for column in columns:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le  # エンコーダを保存
    return df, label_encoders

def simulate_price_increase(data, high_price_categories, label_encoders, price_increase=1.1):
    """
    高単価商品の価格を一定割合増加させたシミュレーションを行う関数。
    data: シミュレーション対象のデータフレーム
    high_price_categories: 高単価商品カテゴリのリスト
    label_encoders: カテゴリカル変数のラベルエンコーダ
    price_increase: 価格増加の倍率（デフォルトは10%増）
    """
    simulation_data = data.copy()
    for category in high_price_categories:
        if category in label_encoders['商品カテゴリ'].classes_:
            category_label = label_encoders['商品カテゴリ'].transform([category])[0]
            simulation_data.loc[simulation_data['商品カテゴリ'] == category_label, '商品単価'] *= price_increase
    return simulation_data

def compare_actual_and_predicted_sales(df, model, learn_columns):
    """
    実売上と予測売上の比較を行い、合計を表示する関数。
    df: 実売上と予測のためのデータフレーム
    model: 予測に使うモデル
    learn_columns: 説明変数のリスト
    """
    X_test = df[learn_columns].copy()
    df['予測数量'] = model.predict(X_test)
    df['予測売上'] = df['予測数量'] * df['商品単価']

    total_actual_sales = df['売上'].sum()
    total_predicted_sales = df['予測売上'].sum()

    print(f"実売上の合計: {total_actual_sales:.2f}円")
    print(f"予測売上の合計: {total_predicted_sales:.2f}円")

    return df
```

```python
def date_data_preprocessing(df):
    # --- '取引日時' 列を基に '年', '月', '日', '時間' の新しい列を作成 ---
    df['取引日時'] = pd.to_datetime(df['取引日時'])
    df['年'] = df['取引日時'].dt.year
    df['月'] = df['取引日時'].dt.month
    df['日'] = df['取引日時'].dt.day
    df['時間'] = df['取引日時'].dt.hour
    df['日付'] = pd.to_datetime(df['日付'])
    df.set_index('日付', inplace=True)
    df.drop('取引日時', axis=1, inplace=True)
    return df
```

## 5.分析から考えられる施策のシミュレーション

```python
import pandas as pd
import matplotlib.pyplot as plt

def price_adjustment_simulation(df, price_increase, target_categories, adjustment_time_period):
    """
    指定したカテゴリと時間帯で価格を調整し、調整前後の価格および売上をシミュレーションする関数。

    Parameters:
    df: データフレーム（商品カテゴリ、時間帯、商品単価、数量、売上を含む）
    price_increase: 価格調整の割合（例：1.1は10%の値上げ、0.9は10%の値下げ）
    target_categories: 価格調整対象のカテゴリのリスト
    adjustment_time_period: 調整対象の時間帯のリスト（例：['17:00:00', '18:00:00']）

    Returns:
    scenario_df: 調整前後の価格と売上を含むデータフレーム
    """

    # データをコピーしてシミュレーション用のデータフレームを作成
    scenario_df = df.copy()

    # 元の価格を保存
    scenario_df['元の価格'] = scenario_df['商品単価']

    # 対象となるカテゴリと時間帯で価格を調整する
    scenario_df['シミュレーション後の商品単価'] = scenario_df.apply(
        lambda row: row['商品単価'] * price_increase
        if (row['商品カテゴリ'] in target_categories and row['時間帯'] in adjustment_time_period)
        else row['商品単価'],
        axis=1
    )

    # 元の売上とシミュレーション後の売上を計算
    scenario_df['元の売上'] = scenario_df['数量'] * scenario_df['元の価格']
    scenario_df['シミュレーション後の売上'] = scenario_df['数量'] * scenario_df['シミュレーション後の商品単価']

    # 価格差を計算
    scenario_df['価格差'] = scenario_df['シミュレーション後の商品単価'] - scenario_df['元の価格']

    return scenario_df
```

```python
# サンプルのデータフレーム（df）を使ってシミュレーションを実行

# シミュレーションの設定
price_increase = 1.2  # 20%の値上げ
target_categories = ['ガイド', '撮影料']  # 価格調整対象のカテゴリ
adjustment_time_period = ['17:00:00', '18:00:00']  # 夕方の時間帯で価格調整

# シミュレーションの実行
scenario_A = price_adjustment_simulation(sales_df, price_increase, target_categories, adjustment_time_period)

# # シミュレーション結果の確認
# plot_price_simulation_results(scenario_A)

# # 調整前後の価格と売上を表示（サンプル）
# scenario_A[['商品カテゴリ', '時間帯', '元の価格', 'シミュレーション後の商品単価', '元の売上', 'シミュレーション後の売上']].head(10)
```

```python
# Define the COVID events date ranges
covid_events = [
    {"event": "第1回緊急事態宣言", "start": "2020-04-07", "end": "2020-05-25"},
    {"event": "Go To トラベルキャンペーン", "start": "2020-07-22", "end": "2020-12-28"},
    {"event": "第2回緊急事態宣言", "start": "2021-01-07", "end": "2021-03-21"},
    {"event": "第3回緊急事態宣言", "start": "2021-04-25", "end": "2021-06-20"},
    {"event": "コロナ禍のオリンピックにおける政府対応", "start": "2021-07-12", "end": "2021-09-30"}
]

# Convert the event start and end dates to datetime
for event in covid_events:
    event['start'] = pd.to_datetime(event['start'])
    event['end'] = pd.to_datetime(event['end'])

# Determine the earliest and latest dates of the events
start_of_covid = min(event['start'] for event in covid_events)
end_of_covid = max(event['end'] for event in covid_events)
```
